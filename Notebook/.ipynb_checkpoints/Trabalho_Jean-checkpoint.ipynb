````python
# Importando as bibliotecas.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Carregando as bases gerada pelo AgrawalGenerator no MOA.
df_Hoeffding_Agrawal = pd.read_csv('../AgrawalGenerator/HoeffdingTree_38,71s.txt')
df_Naive_Agrawal = pd.read_csv('../AgrawalGenerator/NaiveBayes_2,91s.txt')
df_kNN_Agrawal = pd.read_csv('../AgrawalGenerator/kNN_4m7s.txt')
df_Oza_Agrawal = pd.read_csv('../AgrawalGenerator/OzaBagAdwin_2m56s.txt')
````

````python
# Plotando o gráfico de acertos AgrawalGenerator com seus classificadores.
performance = (df_Hoeffding_Agrawal['classifications correct (percent)'].max(), df_kNN_Agrawal['classifications correct (percent)'].max(), df_Naive_Agrawal['classifications correct (percent)'].max(), df_Oza_Agrawal['classifications correct (percent)'].max())
objects = ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin')
y_pos = np.arange(len(objects))

# plt.figure(figsize=(20,10))
plt.xticks(y_pos, objects)
plt.ylabel('Accuracy')
plt.title('Algorithms x Accuracy')

rects1 = plt.bar(y_pos, performance, align='center', alpha=0.5)

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., height,
                '%.2f' % float(height),
                ha='center', va='bottom')

autolabel(rects1)
 
plt.show()
````

````python
# Plotando o gráfico entre Algoritmos x Evalution Time Agrawal.
evaluation = (df_Hoeffding_Agrawal['evaluation time (cpu seconds)'].max(), df_kNN_Agrawal['evaluation time (cpu seconds)'].max(), df_Naive_Agrawal['evaluation time (cpu seconds)'].max(), df_Oza_Agrawal['evaluation time (cpu seconds)'].max())
objects = ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin')
y_pos = np.arange(len(objects))
plt.figure(figsize=(20,10))
rect1 = plt.barh(y_pos, evaluation, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Time Elapsed')
plt.title('Algoritmos x Evaluation Time')


plt.show()
````

````python
# Plotando o gráfico entre Algoritmos x Model Cost (RAM-HOURS) Agrawal.
evaluation = (df_Hoeffding_Agrawal['evaluation time (cpu seconds)'].max(), df_kNN_Agrawal['evaluation time (cpu seconds)'].max(), df_Naive_Agrawal['evaluation time (cpu seconds)'].max(), df_Oza_Agrawal['evaluation time (cpu seconds)'].max())
objects = ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin')
y_pos = np.arange(len(objects))
plt.figure(figsize=(20,10))
rect1 = plt.barh(y_pos, evaluation, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Time Elapsed')
plt.title('Algoritmos x Evaluation Time')


plt.show()
````

````python
# Plotando o gráfico entre Algoritmos x Model Cost RAM Agrawal.

cost = (df_Hoeffding_Agrawal['model cost (RAM-Hours)'].max(), df_kNN_Agrawal['model cost (RAM-Hours)'].max(), df_Naive_Agrawal['model cost (RAM-Hours)'].max(), df_Oza_Agrawal['model cost (RAM-Hours)'].max())
objects = ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin')
y_pos = np.arange(len(objects))

plt.figure(figsize=(20,10))

rect1 = plt.barh(y_pos, cost, align='center', alpha=0.5)

plt.yticks(y_pos, objects)
plt.xlabel('Time Elapsed')
plt.title('Algoritmos x Cost RAM')
    
    
plt.show()
````

````python
# Plotando o gráfico de acurácia por instancias AgrawalGenerator.
plt.figure(figsize=(20,10))
plt.plot(df_Hoeffding_Agrawal['learning evaluation instances'], df_Hoeffding_Agrawal['classifications correct (percent)'], label='HoeffdingTree')
plt.plot(df_Naive_Agrawal['learning evaluation instances'], df_Naive_Agrawal['classifications correct (percent)'], label='NaiveBayes')
plt.plot(df_kNN_Agrawal['learning evaluation instances'], df_kNN_Agrawal['classifications correct (percent)'], label='kNN')
plt.plot(df_Oza_Agrawal['learning evaluation instances'], df_Oza_Agrawal['classifications correct (percent)'], label='OzaBagAdwin')
plt.axvspan(180000, 220000, color='red', alpha=0.5)
plt.axvspan(380000, 420000, color='red', alpha=0.5)
plt.axvspan(580000, 620000, color='red', alpha=0.5)
plt.axvspan(780000, 820000, color='red', alpha=0.5)
plt.xlim(0, 1000000)
plt.ylim(40, 100)

# Add the legend with some customizations.
legend = plt.legend(loc='lower left', shadow=True)

# The frame is matplotlib.patches.Rectangle instance surrounding the legend.
frame = legend.get_frame()
frame.set_facecolor('0.90')

# Set the fontsize
for label in legend.get_texts():
    label.set_fontsize('large')
    
for label in legend.get_lines():
    label.set_linewidth(1.5)

# Plot chart.
plt.show()
````

````python
# data to plot evaluation time
Agrawal_Generator = (df_Hoeffding_Agrawal['model cost (RAM-Hours)'].max(), df_kNN_Agrawal['model cost (RAM-Hours)'].max(), df_Naive_Agrawal['model cost (RAM-Hours)'].max(), df_Oza_Agrawal['model cost (RAM-Hours)'].max())
Asset_Generator = (df_Hoeffding_Asset['model cost (RAM-Hours)'].max(), df_kNN_Asset['model cost (RAM-Hours)'].max(), df_Naive_Asset['model cost (RAM-Hours)'].max(), df_Oza_Asset['model cost (RAM-Hours)'].max())
SEA_Generator = (df_Hoeffding_SEA['model cost (RAM-Hours)'].max(), df_kNN_SEA['model cost (RAM-Hours)'].max(), df_Naive_SEA['model cost (RAM-Hours)'].max(), df_Oza_SEA['model cost (RAM-Hours)'].max())

n_groups = 4

# create plot
index = np.arange(n_groups)
bar_width = 0.30
opacity = 0.8
plt.figure(figsize=(18,8))

rects1 = plt.bar(index, Agrawal_Generator, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Agrawal_Generator')
 
rects2 = plt.bar(index + bar_width, Asset_Generator, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Asset_Generator')

rects3 = plt.bar(index + bar_width + bar_width, SEA_Generator, bar_width,
                 alpha=opacity,
                 color='y',
                 label='SEA_Generator')
 
plt.xlabel('Algorithms')
plt.ylabel('RAM Used')
plt.title('RAM x Algorithms')
plt.xticks(index + bar_width, ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin'))
plt.legend()

 
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    count=0
    for rect in rects:
        height = rect.get_height()
        if(count == 2):
            plt.text(rect.get_x() + rect.get_width()/2., height,
                '%.2e' % float(height),
                ha='center', va='bottom')
            count = 0
        else:
            plt.text(rect.get_x() + rect.get_width()/2., height,
                '%f' % float(height),
                ha='center', va='bottom')
            count+=1


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
````
