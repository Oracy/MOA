# Informações dos testes no MOA

Algoritmos Geradores para classificação: 
    * Agrawal
    * Assets
    * SEA

Algoritmos Classificadores:
    * HoeffdingTree
    * kNN
    * Naive Bayes
    * OzaBagAdwin
    
Este trabalho consiste em gerar 1.000.000 de instâncias através de geradores sintéticos de dados citados acima utilizando 4 mudanças de conceitos dentro destas instâncias, sendo feita as mudanças a cada 200.000 instâncias.

Após esta criação dos dados sintéticos utilizar os quatros classificadores citados acima

````python
# Importando as bibliotecas.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
# Carregando as bases gerada pelo AgrawalGenerator no MOA.
df_Hoeffding_Agrawal = pd.read_csv('../AgrawalGenerator/HoeffdingTree_38,71s.txt')
df_Naive_Agrawal = pd.read_csv('../AgrawalGenerator/NaiveBayes_2,91s.txt')
df_kNN_Agrawal = pd.read_csv('../AgrawalGenerator/kNN_4m7s.txt')
df_Oza_Agrawal = pd.read_csv('../AgrawalGenerator/OzaBagAdwin_2m56s.txt')
````

````python
# Plotando o gráfico de acurácia máxima do AgrawalGenerator com seus classificadores.
performance = (df_Hoeffding_Agrawal['classifications correct (percent)'].max(), df_kNN_Agrawal['classifications correct (percent)'].max(), df_Naive_Agrawal['classifications correct (percent)'].max(), df_Oza_Agrawal['classifications correct (percent)'].max())
objects = ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin')
y_pos = np.arange(len(objects))

plt.xticks(y_pos, objects)
plt.ylabel('Accuracy')
plt.title('Algorithms x Accuracy')

rects1 = plt.bar(y_pos, performance, align='center', alpha=0.20, color='r')

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., height,
                '%.2f' % float(height),
                ha='center', va='bottom')

# Definir as cores e o gráfico em destaque        
rects1[3].set_color('g')
rects1[3].set_alpha(0.7)

plt.axhline(95, color='r', linestyle='dashed', linewidth=1, label='1211132')

autolabel(rects1)
 
plt.show()
````

````python
x1 = df_Hoeffding_Agrawal['evaluation time (cpu seconds)'].max()
x2 = df_kNN_Agrawal['evaluation time (cpu seconds)'].max()
x3 = df_Naive_Agrawal['evaluation time (cpu seconds)'].max()
x4 = df_Oza_Agrawal['evaluation time (cpu seconds)'].max()
y1 = df_Hoeffding_Agrawal['model cost (RAM-Hours)'].max()
y2 = df_kNN_Agrawal['model cost (RAM-Hours)'].max()
y3 = df_Naive_Agrawal['model cost (RAM-Hours)'].max()
y4 = df_Oza_Agrawal['model cost (RAM-Hours)'].max()
z1 = df_Hoeffding_Agrawal['classifications correct (percent)'].max()
z2 = df_kNN_Agrawal['classifications correct (percent)'].max()
z3 = df_Naive_Agrawal['classifications correct (percent)'].max()
z4 = df_Oza_Agrawal['classifications correct (percent)'].max()

labels = ['HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin']
fig = plt.figure()
ax = Axes3D(fig)

scatter1 = ax.scatter(x1, y1, z1, color='g')
scatter2 = ax.scatter(x2, y2, z2, color='b')
scatter3 = ax.scatter(x3, y3, z3, color='y')
scatter4 = ax.scatter(x4, y4, z4, color='r')
ax.set_xlabel('evaluation time (cpu seconds)')
ax.set_ylabel('model cost (RAM-Hours)')
ax.set_zlabel('classifications correct (percent)')
# ax.set_ylim(0.00000, 0.00025)
ax.set_zlim3d(40, 100)

ax.legend([scatter1, scatter2, scatter3, scatter4], ['HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin'], loc='upper left')

plt.show()
````

````python
# Plotando o gráfico entre Algoritmos x Evalution Time Agrawal.
evaluation = (df_Hoeffding_Agrawal['evaluation time (cpu seconds)'].max(), df_kNN_Agrawal['evaluation time (cpu seconds)'].max(), df_Naive_Agrawal['evaluation time (cpu seconds)'].max(), df_Oza_Agrawal['evaluation time (cpu seconds)'].max())
objects = ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin')
y_pos = np.arange(len(objects))
plt.figure(figsize=(20,10))
rect1 = plt.barh(y_pos, evaluation, align='center', alpha=0.2, color='r')
plt.yticks(y_pos, objects)
plt.xlabel('Time Elapsed')
plt.title('Algoritmos x Evaluation Time')

rect1[2].set_color('g')
rect1[2].set_alpha(0.7)


plt.show()
````

````python
# Plotando o gráfico entre Algoritmos x Model Cost RAM Agrawal.

cost = (df_Hoeffding_Agrawal['model cost (RAM-Hours)'].max(), df_kNN_Agrawal['model cost (RAM-Hours)'].max(), df_Naive_Agrawal['model cost (RAM-Hours)'].max(), df_Oza_Agrawal['model cost (RAM-Hours)'].max())
objects = ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwin')
y_pos = np.arange(len(objects))

plt.figure(figsize=(20,10))

rect1 = plt.barh(y_pos, cost, align='center', alpha=0.2, color='r')

plt.yticks(y_pos, objects)
plt.xlabel('Model Cost (RAM-HOURS)')
plt.title('Algorithms x Model Cost (RAM-HOURS)')

rect1[2].set_color('g')
rect1[2].set_alpha(0.7)
    
    
plt.show()
````
