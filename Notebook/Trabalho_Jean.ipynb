Este trabalho consiste em gerar 1.000.000 de instâncias em cada gerador sitético informado acima, e 4 mudanças de conceitos, ocorrendo a cada 200.000 instâncias,

````python
# Dados para plotar o Tempo de Processamento
Agrawal_Generator = (df_Hoeffding_Agrawal['evaluation time (cpu seconds)'].max(), df_kNN_Agrawal['evaluation time (cpu seconds)'].max(), df_Naive_Agrawal['evaluation time (cpu seconds)'].max(), df_Oza_Agrawal['evaluation time (cpu seconds)'].max())
Asset_Generator = (df_Hoeffding_Asset['evaluation time (cpu seconds)'].max(), df_kNN_Asset['evaluation time (cpu seconds)'].max(), df_Naive_Asset['evaluation time (cpu seconds)'].max(), df_Oza_Asset['evaluation time (cpu seconds)'].max())
SEA_Generator = (df_Hoeffding_SEA['evaluation time (cpu seconds)'].max(), df_kNN_SEA['evaluation time (cpu seconds)'].max(), df_Naive_SEA['evaluation time (cpu seconds)'].max(), df_Oza_SEA['evaluation time (cpu seconds)'].max())

n_groups = 4

# Cria o gráfico
index = np.arange(n_groups)
bar_width = 0.30
opacity = 0.8
plt.figure(figsize=(20,10))

rects1 = plt.bar(index, Agrawal_Generator, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Agrawal_Generator')
 
rects2 = plt.bar(index + bar_width, Asset_Generator, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Asset_Generator')

rects3 = plt.bar(index + bar_width + bar_width, SEA_Generator, bar_width,
                 alpha=opacity,
                 color='y',
                 label='SEA_Generator')
 
plt.xlabel('Algorithms')
plt.ylabel('Time Elapsed')
plt.title('Time x Algorithms')
plt.xticks(index + bar_width, ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwim'))
plt.legend()
 
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., height,
                '%f' % float(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
````

````python
# Dados para plotar o Gasto de Memória
Agrawal_Generator = (df_Hoeffding_Agrawal['model cost (RAM-Hours)'].max(), df_kNN_Agrawal['model cost (RAM-Hours)'].max(), df_Naive_Agrawal['model cost (RAM-Hours)'].max(), df_Oza_Agrawal['model cost (RAM-Hours)'].max())
Asset_Generator = (df_Hoeffding_Asset['model cost (RAM-Hours)'].max(), df_kNN_Asset['model cost (RAM-Hours)'].max(), df_Naive_Asset['model cost (RAM-Hours)'].max(), df_Oza_Asset['model cost (RAM-Hours)'].max())
SEA_Generator = (df_Hoeffding_SEA['model cost (RAM-Hours)'].max(), df_kNN_SEA['model cost (RAM-Hours)'].max(), df_Naive_SEA['model cost (RAM-Hours)'].max(), df_Oza_SEA['model cost (RAM-Hours)'].max())

n_groups = 4

# Cria o gráfico
index = np.arange(n_groups)
bar_width = 0.30
opacity = 0.8
plt.figure(figsize=(18,8))

rects1 = plt.bar(index, Agrawal_Generator, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Agrawal_Generator')
 
rects2 = plt.bar(index + bar_width, Asset_Generator, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Asset_Generator')

rects3 = plt.bar(index + bar_width + bar_width, SEA_Generator, bar_width,
                 alpha=opacity,
                 color='y',
                 label='SEA_Generator')
 
plt.xlabel('Algorithms')
plt.ylabel('RAM Used')
plt.title('RAM x Algorithms')
plt.xticks(index + bar_width, ('HoeffdingTree', 'kNN', 'NaiveBayes', 'OzaBagAdwim'))
plt.legend()

 
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    count=0
    for rect in rects:
        height = rect.get_height()
        if(count == 2):
            plt.text(rect.get_x() + rect.get_width()/2., height,
                '%.g' % float(height),
                ha='center', va='bottom')
            count = 0
        else:
            plt.text(rect.get_x() + rect.get_width()/2., height,
                '%f' % float(height),
                ha='center', va='bottom')
            count+=1


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
````
